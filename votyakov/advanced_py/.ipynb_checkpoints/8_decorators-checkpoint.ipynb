```python
#фибонача
def fib(x): 
    if x <= 1: return 1
    else:
        return fib(x-1) + fib(x-2) 
for n in range(10): 
    print(n, fib(n)) # -- неэффективно 
    

from functools import lru_cache, cache

@cache
def fib(x): 
    if x <= 1: return 1
    else:
        return fib(x-1) + fib(x-2) 
for n in range(100): 
    print(n, fib(n)) # -- эффективно
    
#или

@lru_cache(None)
def fib(x): 
    if x <= 1: return 1
    else:
        return fib(x-1) + fib(x-2) 
for n in range(100): 
    print(n, fib(n)) # -- эффективно
```

```python
def adder(x): 
    def g(y): 
        return x+y
    return g
print(adder(15)(30))
```

```python
def kvadrator(func): 
    def g(x): 
        return func(x)**2
    return g

def f(x): 
    return x+2

print(f(10))
print(kvadrator(f)(10))
```

```python
#не используя декораторы 
f = kvadrator(f) #"урашаем квадратором"
print(f(10))
```

```python
#декораторы 
def kvadrator(func): 
    def g(x): 
        return func(x)**2
    return g

@kvadrator
def f(x): 
    return x+2

print(f(10))
```

```python
def my_decorator(func):
    
    def inside_func():
        print('Привет, я - вступление из декоратора')
        func()
        print('Все, это уже выход из декоратора')
        
    return inside_func
        
@my_decorator 
def f():
    print('Привет, я оригинальная функция!')
    
#f = my_decorator(f) - идентично

f()
```

```python
#двойной декоратор

def my_decorator(func):
    def inside_func():
        print('Привет, я - вступление из декоратора')
        func()
        print('Все, это уже выход из декоратора')        
    return inside_func

def doubler(func): 
    def inside_func():
        func()
        func() 
    return inside_func

@doubler
@my_decorator 
def f():
    print('Привет, я оригинальная функция!')

f()
```

```python
@doubler
def f():
    print('Привет, я оригинальная функция!')

f()
```

```python
@my_decorator 
@doubler
def f():
    print('Привет, я оригинальная функция!')

f()
```

```python
#собственное кеширование

def my_cacher(func): 
    D = dict() #создается 1 словарь (не пересоздается)
    def inside_func(x): 
        if x not in D:
            D[x] = func(x) 
        return D[x]
    return inside_func

@my_cacher
def fib(x): 
    if x <= 1: 
        return 1
    else: 
        return fib(x-1) + fib(x-2)
    
for n in range(100):
    print(n, fib(n))
```

```python
import time

def timer(func):
    def inside_func(*args, **kwargs): 
        st = time.time()
        func(*args, **kwargs)
        fin = time.time()
        print(f'Функция {func.__name__} проработала {fin-st:.10f} секунд')
    return inside_func

@timer
def f(x): 
    k = 0
    for i in range(x): 
        for j in range(x):
            if (i + j) % 100000 == 0:
                k += 1
                
    print(k)

f(1000)
f(2000)
```

```python
#с ретерном

import time

def timer(func):
    def inside_func(*args, **kwargs): 
        st = time.time()
        result = func(*args, **kwargs)
        fin = time.time()
        print(f'Функция {func.__name__} проработала {fin-st:.10f} секунд')
        
        return result 
    return inside_func

@timer
def f(x): 
    k = 0
    for i in range(x): 
        for j in range(x):
            if (i + j) % 100000 == 0:
                k += 1
                
    return k

res = f(1000)
print(res)
```
